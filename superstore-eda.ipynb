**Sales Trend over time (monthly , yearly)**

```python
monthly_sales = superstore.groupby(['Order Date'] , as_index=False).sum().set_index('Order Date')

monthly_sales = monthly_sales.resample('M').sum()



plt.figure(figsize=(25,8))
plt.plot(monthly_sales['Sales'])
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
```

```python
monthly_sales = superstore.groupby(['Order Date'] , as_index=False).sum().set_index('Order Date')

monthly_sales = monthly_sales.resample('M').sum()



plt.figure(figsize=(25,8))
plt.plot(monthly_sales['Profit'])
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.title('Monthly Profit Trend')
```

```python
yearly_sales = monthly_sales.resample('Y').sum()



plt.figure(figsize=(25,8))
plt.plot(yearly_sales['Sales'])
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Yearly Sales Trend')
```

**Regions & Places generates the most Sales**

```python
superstore_places = superstore[ ['Country' , 'City'	 , 'State' , 'Region'] ]
```

```python
superstore_places.nunique()
```

```python
superstore_places = superstore[ [ 'City', 'State' , 'Region' , 'Sales' , 'Profit'] ]
```

```python
superstore_places.head(5)
```

```python
grouped_data = superstore_places.groupby(['Region'] , as_index=False).sum()
grouped_data.sort_values(by = 'Sales' , ascending=False , inplace = True )

plt.figure(figsize=(10,5))
plt.bar(grouped_data['Region'] , grouped_data['Sales'] , align='center')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales Generated by States')
```

```python
grouped_data = superstore_places.groupby(['Region'] , as_index=False).sum()
grouped_data.sort_values(by = 'Profit' , ascending=False , inplace = True )

plt.figure(figsize=(10,5))
plt.bar(grouped_data['Region'] , grouped_data['Profit'] , align='center')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit Generated by State')
```

```python
grouped_data = superstore_places.groupby(['State'] , as_index=False).sum()
grouped_data.sort_values(by = 'Sales' , ascending=False , inplace = True )

plt.figure(figsize=(22,10))
plt.bar(grouped_data['State'] , grouped_data['Sales'] , align='center')
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('Sales Generated by State')
plt.xticks(rotation = 90)
```

```python
grouped_data = superstore_places.groupby(['City'] , as_index=False).sum()
grouped_data.sort_values(by = 'Sales' , ascending=False , inplace = True )

top_5_cites = grouped_data.head()

plt.figure(figsize=(10,10))

plt.bar(top_5_cites['City'] , top_5_cites['Sales'] , align='center')
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('top 5 cites by Sales')
```

```python
grouped_data = superstore_places.groupby(['City'] , as_index=False).sum()
grouped_data.sort_values(by = 'Profit' , ascending=False , inplace = True )

top_5_cites = grouped_data.head()

plt.figure(figsize=(10,10))

plt.bar(top_5_cites['City'] , top_5_cites['Profit'] , align='center')
plt.xlabel('City')
plt.ylabel('Profit')
plt.title('top 5 cites by Profit')
```
